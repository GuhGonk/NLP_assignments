```python
# We will use Glove, a pre-trained word embedding vectors.
urlretrieve("http://nlp.stanford.edu/data/glove.6B.zip", filename="glove.6B.zip")
zf = zipfile.ZipFile('glove.6B.zip')
zf.extractall()
zf.close()
```

```python
# Glove comes in multiple different formats based on different dimension legnth. We will use 50 (Not 100) dimension Glove vector.
glove_dict = dict()
f = open('glove.6B.50d.txt', encoding="utf8")

for line in f:
    word_vector = line.split()
    word = word_vector[0]
    word_vector_arr = np.asarray(word_vector[1:], dtype='float32')
    glove_dict[word] = word_vector_arr
f.close()
```

```python
# For each word which exists in the file, it is represented with 50 different real numbers.
# print out the embedding vector for the word 'pizza'.
print (glove_dict['pizza'])
```

```python
## We will work on generating "sentence" embedding (not word embedding).

# Assume that you have 10 different sentences and you want to compare these to each other.

# Convert each sentence to a vector (in the same vector length) and perform such as cosine similarity:
  # How to convert each sentence to a sentence embedding vector?
  # Compute the average of word embedding vectors per sentence (These words exist in that sentence).

# Note that cosine similarity tutorial in the past was based on word-level tf-idf vectors & not related to embedding.
```

```python
# For OOV (the words which exist in a data set but do not exist in the Glove vector), generate a 50-length of zero embedding vector.
embedding_dim = 50
zero_vector = np.zeros(embedding_dim)
print (zero_vector)
```

```python
# convert each word into embedding vector & convert OOV into zero embedding vector.
# Compute the average of the word embedding vectors -> sentence embedding vector.

def calculate_sentence_vector(sentence):
  return sum([glove_dict.get(word, zero_vector) for word in sentence])/len(sentence)
```

## (2) For each sentence within the original text (news_contents), perform word-tokenize. Also, apply pre-processings such as lower-casing & removing stop-words.

```python
# word-tokenize
data['sentences'] = data['news_contents'].apply(sent_tokenize)

def word_tk(sentences):
    return [word_tokenize(sentence) for sentence in sentences]

data['sentences_tk'] = data['sentences'].apply(word_tk)

# pre-processing

def preprocess_sentence(sentence):
  # words only, lower case
  sentence = [re.sub(r'[^a-zA-z\s]', '', word).lower() for word in sentence]
  # remove stop-words
  return [word for word in sentence if word not in stop_words and word]

def pre_proc_apply(sentences):
  return [preprocess_sentence(sentence) for sentence in sentences]

data['sentences_tk'] = data['sentences_tk'].apply(pre_proc_apply)

print(data.shape)
data.head()
```

```python
# We will use "news_contents".
# sentence tokenization -> "Setences" column (multiple sentences for each row of "news_contents").
#                       -> apply word-tokenization & pre-processings -> "tokenized_sentences" column.
```

```python
# For OOV (the words which exist in a data set but do not exist in the Glove vector), generate a 50-length of zero embedding vector.
embedding_dim = 50
zero_vector = np.zeros(embedding_dim)
```

## (3) Convert each word into embedding vector, then generate sentence-level embedding vectors.

*   Words that are present within the pre-trained Glove -> convert the words into embedding vectors.
*   Words that are not present within the Glove (OOV) & other cases (i.e. sentence becomes 0 after removing stop-words) -> convert these into zero vectors.



```python
# convert each word into embedding vector & convert OOV into zero embedding vector & others such as a sentence has 0 length after removing stop-words.
# Compute the average of the word embedding vectors -> sentence embedding vector.

def calculate_sentence_vector(sentence):
  if len(sentence) != 0:
    return sum([glove_dict.get(word, zero_vector)
                  for word in sentence])/len(sentence)
  else:
    return zero_vector

# data['sentences_emb'] = data['sentences_tk'].apply(lambda x: [calculate_sentence_vector(sentence) for sentence in data['sentences_tk']])

def sentences_to_vectors(sentences):
    return [calculate_sentence_vector(sentence) for sentence in sentences]

data['sentences_emb'] = data['sentences_tk'].apply(sentences_to_vectors)

print(data.shape)
data.head()
```

## (4) perform cosine similiarity among all the sentences (embedded vectors) for each news article.

```python
# perform consine similiarity for all the available sentences per article.

def similarity_matrix(sentence_embedding):
  sim_mat = np.zeros([len(sentence_embedding), len(sentence_embedding)])
  for i in range(len(sentence_embedding)):
      for j in range(len(sentence_embedding)):
        sim_mat[i][j] = cosine_similarity(sentence_embedding[i].reshape(1, embedding_dim),
                                          sentence_embedding[j].reshape(1, embedding_dim))[0,0]
  return sim_mat

data['SimMatrix'] = data['sentences_emb'].apply(similarity_matrix)

print (data.shape)
data.head()
```

```python
# PageRank computes a ranking of the nodes in a network graph based on the structure of the incoming links.
# It was originally designed as an algorithm to rank web pages.

def calculate_score(sim_matrix):
    nx_graph = nx.from_numpy_array(sim_matrix)
    scores = nx.pagerank(nx_graph)
    return scores
```

## (7) Based on the highest score values, choose 3 highest scored sentences and consider these sentences as a summary output per article.

```python
def ranked_sentences(sentences, scores, n=3):
    # sort the score from the highest to the lowest.
    top_scores = sorted(((scores[i],s) for i,s in enumerate(sentences)), reverse=True)
    # choose the corresponding sentences.
    top_n_sentences = [sentence for score,sentence in top_scores[:n]]
    # put these sentences into a single paragraph.
    return " ".join(top_n_sentences)


data['summary'] = data.apply(lambda x: ranked_sentences(x.sentences, x.Final_Score), axis=1)

print (data.shape)
data.head()
```
